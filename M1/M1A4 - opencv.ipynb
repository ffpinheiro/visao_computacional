```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread("python144x144.png")
```

```python
print(type(image))
```

```python
plt.imshow(image)
```

```python
gain = np.clip(image /255 * 1.001,0,1)
plt.imshow(gain)
```

```python
gain2 = np.clip(image /255 * 1.901,0,1) # adição minha
plt.imshow(gain2)
```

```python
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gain_RGB = np.clip(image_RGB /255 * 1.001,0,1)
plt.imshow(gain_RGB)
```

```python
crop = image_RGB[0:50,0:50]
plt.imshow(crop)
```

```python
crop = image_RGB[20:120,20:120]
plt.imshow(crop)
```

```python
larger = cv2.resize(image_RGB, (int(image.shape[1] * 4), int(image.shape[0] * 4)))
plt.imshow(larger)
```
